### Popularidad de los géneros (box plot)

Se ve cuáles son los géneros más populares (como pop, lógicamente) y los menos populares como música de películas, animé, ópera, etc.

```python
popularidad_genero = train[ ["genre", "popularity"] ]
popularidad_genero

plt.figure(dpi=180)
sns.boxplot(data=popularidad_genero, x="genre", y="popularity", color="#1DB954")
plt.xticks(rotation=90)
plt.xlabel("Géneros", labelpad=15)
plt.ylabel("Popularidad", labelpad=15)
plt.title("Popularidad de los géneros", pad=15)
```

### Volumen de los géneros

Se observa que algunos géneros tienen notoriamente menos volumen que otros (como Soundtrack).

```python
loudness_genero = train[ ["genre", "loudness"] ]
loudness_genero

plt.figure(dpi=180)
sns.boxplot(data=loudness_genero, x="genre", y="loudness", color="#1DB954")
plt.xticks(rotation=90)
plt.xlabel("Géneros", labelpad=15)
plt.ylabel("Volumen", labelpad=15)
plt.title("Volumen de los géneros", pad=15)
```

### Promedio de instrumentalness por género (bar plot)

Algunos géneros son mucho más instrumentales que otros, lo que es esperable en géneros como Soundtrack, electrónica o clásica.

```python
instrumentalness_genero = train[ ["genre", "instrumentalness"] ]
instrumentalness_genero = instrumentalness_genero.groupby(["genre"]).mean().reset_index()

plt.figure(dpi=180)
sns.barplot(data=instrumentalness_genero, x="genre", y="instrumentalness", color="#1DB954")
plt.xticks(rotation=90)
plt.xlabel("Géneros", labelpad=15)
plt.ylabel("Promedio de instrumentalidad", labelpad=15)
plt.title("Instrumentalidad de los géneros", pad=15)
```

### Tonalidad según el género (heatmap)

Se puede ver qué tonalidades predominan según el género, revelando algunas anomalías como muchas canciones de Soundtrack en D.

```python
key_genero = train[ ["genre", "key"] ]
key_genero = key_genero.pivot_table(index='key', columns='genre', aggfunc=len, fill_value=0)
key_genero_normalizado = (key_genero-key_genero.min())/(key_genero.max()-key_genero.min())

plt.figure(dpi=180)
grafico = sns.heatmap(key_genero_normalizado, square=True, cmap=sns.color_palette("Blues", as_cmap=True))
plt.title("Tonalidad según el género", pad=25)
plt.xlabel("Géneros", labelpad=15)
plt.ylabel("Tonalidad", labelpad=15)
grafico.tick_params(left=False, bottom=False)
```

### Speechiness de los géneros (bar plot)

Se observa claramente cuáles son los géneros más hablados, o con más palabras en sus canciones, como comedia, rap o hip-hop.

```python
speechiness_genero = train[ ["genre", "speechiness"] ]
speechiness_genero = speechiness_genero.groupby(["genre"]).mean().reset_index()

plt.figure(dpi=180)
sns.barplot(data=speechiness_genero, x="genre", y="speechiness", color="#1DB954")
plt.xticks(rotation=90)
plt.xlabel("Géneros", labelpad=15)
plt.ylabel("Speechiness", labelpad=15)
plt.title("Speechiness de los géneros", pad=15)
```

### Energía de los géneros (violin plot)

Se observan las distribuciones de energía en los distintos géneros. Suele coincidir con los géneros con la misma bailabilidad.

```python
energy_genero = train[ ["genre", "energy"] ]
energy_genero

plt.figure(dpi=180)
sns.violinplot(data=energy_genero, x="genre", y="energy", color="#1DB954", scale='width')
plt.xticks(rotation=90)
plt.xlabel("Géneros", labelpad=15)
plt.ylabel("Energía", labelpad=15)
plt.title("Energía de los géneros", pad=15)
```
